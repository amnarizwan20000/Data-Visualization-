```python
def grade(per):
    grade = None
    
    if per >= 80:
        grade = "A+"
    elif per >= 70:
        grade = "A"
    elif per >= 60:
        grade = "B"
    elif per >= 50:
        grade = "C"
    elif per >= 40:
        grade = "D"
    elif per >= 33:
        grade = "E"
    else:
        grade = "Fail"
        
    return grade

grade(28)
```

```python
def grading(sub1,sub2,sub3,sub4,sub5):
    obt = sub1 + sub2 + sub3 + sub4 + sub5
    total = 500
    per = obt / total * 100
    
    grade = None
    
    if per >= 80:
        grade = "A+"
    elif per >= 70:
        grade = "A"
    elif per >= 60:
        grade = "B"
    elif per >= 50:
        grade = "C"
    elif per >= 40:
        grade = "D"
    elif per >= 33:
        grade = "E"
    else:
        grade = "Fail"
    return total, obt, per, grade

df1[["total",'obtained','percentage','grade']]= df1.loc[:, "s1":"s5"].apply(lambda x:grading(*x), axis=1, result_type="expand")
df1
```
